# Random Forest
I choose the random forest because it's seems that the categorical variable influences more than numerical. Let's do all the precedure: 

1- Apply LabelEncorder in the categoricals 

```python
df_train.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_CryoSleep = LabelEncoder()
label_econder_transported = LabelEncoder()
```

```python
x=df_train.iloc[:,0:6].values
y=df_train.iloc[:,6].values
```

```python
x[:,0] = label_encoder_CryoSleep.fit_transform(x[:,0])
y=label_econder_transported.fit_transform(y)
```

```python
df=df_test
x_test = df.iloc[:,0:6].values
```

```python
# Scaling the data 
from sklearn.preprocessing import StandardScaler
scaler_census = StandardScaler()
x = scaler_census.fit_transform(x)
```

```python
scaler_test= StandardScaler()
x_test = scaler_test.fit_transform(x_test)
```

```python
import pickle
with open('titanic.pkl',mode='wb') as f:
  pickle.dump([x, y,x_test], f)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
random_forest.fit(x,y)
```

```python
previsoes = random_forest.predict(x_test)
previsoes
```

```python
previsoes_back = label_econder_transported.inverse_transform(previsoes)
previsoes_back
```

```python
sub = df_test
```

```python
sub
```

```python
sub.drop(['CryoSleep','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck',],axis=1,inplace=True)
```

```python
sub['Transported'] = pd.DataFrame(previsoes_back)
```

```python
sub
```

```python
sub.to_csv('submission2.csv', index=False)
```
